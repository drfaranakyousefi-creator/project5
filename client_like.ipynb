```python
# ─── همه کدها در یک cell ─────────────────────────────────────────────────────

import importlib
import new_dataset
import client_net
import transmitter_simulation
import Train_simulation
import plot
import matplotlib.pyplot as plt

importlib.reload(new_dataset)
importlib.reload(client_net)
importlib.reload(transmitter_simulation)
importlib.reload(Train_simulation)
importlib.reload(plot)

from Train_simulation import CAT

# ─── مسیر dataset ────────────────────────────────────────────────────────────
# اگه روی Google Colab هستی:
# from google.colab import drive
# drive.mount('/content/drive')
# chartevents_path = "/content/drive/MyDrive/split_learning/CHARTEVENTS.csv"

# اگه local هستی:
chartevents_path = "./CHARTEVENTS.csv"

# ─── ساخت و آموزش مدل carevue ───────────────────────────────────────────────
first_CAT = CAT(
    seq_len          = 50,
    dataset_name     = 'carevue',
    batch_size       = 128,
    test_size        = 0.2,
    target           = 'spO2',
    d_latent         = 8,
    h                = 4,
    dropout          = 0.1,
    cap_in_dim       = 10,
    lr               = 0.001,
    chartevents_path = chartevents_path,
)

history_first = first_CAT.fit(10)
plot.plot_history(history_first)

# ─── ساخت مدل metavision ─────────────────────────────────────────────────────
second_CAT = CAT(
    seq_len          = 50,
    dataset_name     = 'metavision',
    batch_size       = 128,
    test_size        = 0.2,
    target           = 'spO2',
    d_latent         = 8,
    h                = 4,
    dropout          = 0.1,
    cap_in_dim       = 10,
    lr               = 0.001,
    chartevents_path = chartevents_path,
)

# ─── انتقال دانش از carevue به metavision ────────────────────────────────────
second_CAT.get_knowledge(first_CAT)

# ─── آموزش مدل metavision ────────────────────────────────────────────────────
history_second = second_CAT.fit(10)
plot.plot_history(history_second)

# ─── مقایسه عددی نتایج ───────────────────────────────────────────────────────
print("=== carevue ===")
print(f"best test MSE : {min(history_first['loss_test']):.6f}")
print(f"best test R²  : {max(history_first['r2_test']):.4f}")

print("\n=== metavision (با انتقال دانش) ===")
print(f"best test MSE : {min(history_second['loss_test']):.6f}")
print(f"best test R²  : {max(history_second['r2_test']):.4f}")

# ─── رسم مقایسه‌ای دو مدل ────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(history_first['loss_test'],  label='carevue test',    color='blue')
axes[0].plot(history_second['loss_test'], label='metavision test', color='orange')
axes[0].set_title('Test MSE مقایسه')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('MSE')
axes[0].legend()
axes[0].grid(True)

axes[1].plot(history_first['r2_test'],  label='carevue test',    color='blue')
axes[1].plot(history_second['r2_test'], label='metavision test', color='orange')
axes[1].set_title('Test R² مقایسه')
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('R²')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```
